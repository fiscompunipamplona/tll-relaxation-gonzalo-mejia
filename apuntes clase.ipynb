## Metodo_Relajacion

```python
from pylab import plot,exp,show,sin,arccos
n=20
x0=1.5
x=[]
x.append(x0)
for i in range(n):
    x0=2-exp(-x0)
    x.append(x0)
plot(x)
show()
print(x0)
```

```python
n=100
x0=0.5
x=[]
x.append(x0)
for i in range(n):
    x0=exp(1-(x0*x0))
    x.append(x0)
plot(x)
show()
print(x0)
```

```python
import math 
n=100
x0=0.7
x=[]
x.append(x0)
for i in range(n):
    x0=(1-math.log(x0))**0.5
    x.append(x0)
plot(x)
show()
print(x0)
```

```python
n=100
x0=0.5
x=[]
x.append(x0)
for i in range(n):
    x0=0.5*math.asin(x0-x0*x0)
    x.append(x0)
plot(x)
show()
print(x0)
```

## Ejemplo Ferromag.

```python
from math import tanh,cosh
from numpy import linspace
from pylab import plot,show,ylim,xlabel,ylabel
```

```python
# Constantes
Tmax = 2.0
points = 1000
accuracy = 1e-6

# Configurando listas para graficar
y = []
temp = linspace(0.01,Tmax,points)

# Bucle Temperatura
for T in temp:
    m1 = 1.0
    error = 1.0
    
    # Bucle hasta que el error sea lo suficiente pequeo
    while error>accuracy:
        m1,m2 = tanh(m1/T),m1
        error = abs((m1-m2)/(1-T*cosh(m2/T)**2))
    y.append(m1)

# Hacer la grafica
plot(temp,y)
ylim(-0.1,1.1)
xlabel("Temperatura")
ylabel("Magnetizacion")
show()
```
